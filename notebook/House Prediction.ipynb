```python
import pandas as pd
import numpy as np 
```

```python
df=pd.read_csv('Housing.csv')
```

```python
df.head()
```

```python
#Data Modification 
```

```python
df['mainroad']=np.where(df['mainroad'].str.contains('no'),0,1)
```

```python
df['guestroom']=np.where(df['guestroom'].str.contains('no'),0,1)
df['basement']=np.where(df['basement'].str.contains('no'),0,1)
df['hotwaterheating']=np.where(df['hotwaterheating'].str.contains('no'),0,1)
df['airconditioning']=np.where(df['airconditioning'].str.contains('no'),0,1)
df['prefarea']=np.where(df['prefarea'].str.contains('no'),0,1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder()
```

```python
encoded=encoder.fit_transform(df[['furnishingstatus']])
```

```python
df1=pd.DataFrame(encoded.toarray(),columns=['Furnished','Semi-Furnished','Unfurnished'])
df1
```

```python
df1=df1.drop(['Unfurnished'],axis=1)
```

```python
df1
```

```python
df=df.drop('furnishingstatus',axis=1)
```

```python
df1,df
```

```python
df
```

```python
df1
```

```python
df_org=pd.concat([df,df1],axis=1)
```

```python
df_org
```

```python
X=df_org.iloc[:,2:]
```

```python
X
```

```python
y=df_org.iloc[:,:1]
y
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train,X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```
